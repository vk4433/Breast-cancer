```python
from sklearn.datasets import load_breast_cancer
cancer_dataset = load_breast_cancer()
```

```python
data = pd.DataFrame(np.c_[cancer_dataset['data'],cancer_dataset['target']],
             columns = np.append(cancer_dataset['feature_names'], ['target']))
```

# Data Preprocessing


```python
x = data.drop("target",axis=1)
y = data["target"]
```

# feature importances

```python
from sklearn.ensemble import GradientBoostingClassifier
g_c = GradientBoostingClassifier()
g_c.fit(x,y)
print(g_c.feature_importances_)
```

```python
vis = pd.Series(g_c.feature_importances_,index=x.columns)
vis.nlargest(20).plot(kind='barh')
plt.title("feature_importances")
plt.show()
```

Splitting  dataset


```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=5)
```

# Feature Scaling


```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x_train_sc = ss.fit_transform(x_train)
x_test_sc = ss.transform(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

# Support Vector Classifier


```python
from sklearn.svm import SVC
s_c = SVC()
s_c.fit(x_train,y_train)
s_c_pre = s_c.predict(x_test)
accuracy_score(y_test,s_c_pre)
```

```python
# by StandardScaler data
s_c2 = SVC()
s_c2.fit(x_train_sc,y_train)
s_c_sc_pre = s_c.predict(x_test_sc)
accuracy_score(y_test,s_c_sc_pre)
```

# Logistic Regression


```python
from sklearn.linear_model import LogisticRegression
lr_classifier = LogisticRegression(random_state = 51, penalty = 'l2')
lr_classifier.fit(x_train, y_train)
y_pred_lr = lr_classifier.predict(x_test)
accuracy_score(y_test, y_pred_lr)
```

```python
# by StandardScaler data
from sklearn.linear_model import LogisticRegression
lr_classifier2 = LogisticRegression(random_state = 51, penalty = 'l2')
lr_classifier2.fit(x_train_sc, y_train)
y_pred_lr_sc = lr_classifier.predict(x_test_sc)
accuracy_score(y_test, y_pred_lr_sc)
```

# K – Nearest Neighbor Classifier


```python
from sklearn.neighbors import KNeighborsClassifier
knn_c = KNeighborsClassifier(n_neighbors=5,metric="euclidean",p=2)
knn_c.fit(x_train,y_train)
knn_pr = knn_c.predict(x_test)
accuracy_score(y_test,knn_pr)
```

```python
# by StandardScaler data
from sklearn.neighbors import KNeighborsClassifier
knn_c2 = KNeighborsClassifier(n_neighbors=5,metric="euclidean",p=2)
knn_c2.fit(x_train_sc,y_train)
knn_pr_sc = knn_c.predict(x_test_sc)
accuracy_score(y_test,knn_pr_sc)
```

# Decision Tree Classifier



```python
from sklearn.tree import DecisionTreeClassifier
D_C = DecisionTreeClassifier(criterion="gini",random_state = 51)
D_C.fit(x_train,y_train)
D_c_pre = D_C.predict(x_test)
accuracy_score(y_test,D_c_pre)
```

```python
# by StandardScaler data
from sklearn.tree import DecisionTreeClassifier
D_C2 = DecisionTreeClassifier(criterion="gini",random_state = 51)
D_C2.fit(x_train_sc,y_train)
D_c_pre2 = D_C.predict(x_test_sc)
accuracy_score(y_test,D_c_pre2)
```

# Random Forest Classifier


```python
from sklearn.ensemble import RandomForestClassifier
rf_classifier = RandomForestClassifier(n_estimators = 20, criterion = 'gini', random_state = 51)
rf_classifier.fit(x_train, y_train)
y_pred_rf = rf_classifier.predict(x_test)
accuracy_score(y_test, y_pred_rf)
```

```python
# by StandardScaler data
from sklearn.ensemble import RandomForestClassifier
rf_classifier2 = RandomForestClassifier(n_estimators = 20, criterion = 'entropy', random_state = 51)
rf_classifier2.fit(x_train_sc, y_train)
y_pred_rf_sc = rf_classifier.predict(x_test_sc)
accuracy_score(y_test, y_pred_rf_sc)
```

# confusion matrix

```python
confusion_matrix = confusion_matrix(y_test,y_pred_rf_sc)
confusion_matrix
```

```python
sns.heatmap(confusion_matrix,annot=True)
```

# Classification Report of Model


```python
print(classification_report(y_test,y_pred_rf_sc))
```
